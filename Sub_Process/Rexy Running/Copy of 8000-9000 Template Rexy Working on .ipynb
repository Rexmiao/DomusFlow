```python
import nltk
nltk.download('book')

from nltk.book import text4  # Using text4, which contains a collection of texts from the Gutenberg project

# Extracting unique book titles from the Gutenberg corpus
book_titles = list(set(text4))

# Filtering out non-alphanumeric characters and ensuring titles are capitalized
book_titles = [title.strip().capitalize() for title in book_titles if title.isalnum()]



```

```python
print(len(book_titles))
books_to_scrape = book_titles[8000:9000]
# books_to_scrape = book_titles[1000:2000]
# etc
print(len(books_to_scrape))
print(books_to_scrape)
```

```python
import requests
from googlesearch import search
from bs4 import BeautifulSoup
import os

# def get_summary(book_title):
#     url = f"https://www.sparknotes.com/lit/{book_title}/summary/"
#     response = requests.get(url)
#     # print(response)
#     soup = BeautifulSoup(response.text, 'html.parser')

#     summary_div = soup.find('div', class_='mainTextContent main-container')
#     # print(summary_div)
#     if summary_div:
#         summary = summary_div.text.strip()
#         return summary
#     else:
#         print(f"Error: Summary not found for {book_title}")
#         return None

def get_sparknotes_url(book_title):
    query = f"{book_title} site:sparknotes.com/lit summary"
    for url in search(query, num=5, stop=5, pause=2):
        if 'sparknotes.com/lit' in url and 'summary' in url:
            return url
    return None


def get_summary_from_sparknotes(book_title):
    # Check if the summary file already exists
    filename = f"{output_dir}/{book_title}_summary.txt"
    if os.path.exists(filename):
      print(f"Summary for {book_title} already exists. Reading from file.")
    else:
      sparknotes_url = get_sparknotes_url(book_title)
      print(sparknotes_url)
      if sparknotes_url:
          response = requests.get(sparknotes_url)
          soup = BeautifulSoup(response.text, 'html.parser')
          summary_div = soup.find('div', class_='mainTextContent main-container')
          if summary_div:
              summary = summary_div.text.strip()
              return summary
          else:
              print(f"Error: Summary not found for {book_title}")
              return None
      else:
          print(f"Error: SparkNotes summary not found for {book_title}")
          return None


def save_summary_to_file(book_title, summary, output_dir = "summaries"):
    if summary is not None:  # Check if summary is not None
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        filename = f"{output_dir}/{book_title}_summary.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(summary)
            print(f"Summary for {book_title} saved successfully.")
    # else:
    #     print(f"No summary found for {book_title}. Skipping saving to file.")

# books_to_scrape = ["1984", "ToKillAMockingbird", "TheGreatGatsby"]  # Add more book titles as needed

for book_title in books_to_scrape:
    try:
        output_dir = "summaries"
        summary = get_summary_from_sparknotes(book_title)
        # print(summary)

        save_summary_to_file(book_title, summary,)
    except Exception as e:
        print(f"Error processing {book_title}: {e}")
```

```python
import os
import pandas as pd

# Path to the directory containing the txt files
txt_dir = '/content/summaries/'

# Lists to store the titles and cleaned contents of the txt files
titles = []
contents = []

# Iterate over all files in the directory
for file_name in os.listdir(txt_dir):
    if file_name.endswith('_summary.txt'):
        # Extract the title from the file name
        title = file_name.replace('_summary.txt', '')
        titles.append(title)

        # Read the contents of each txt file
        with open(os.path.join(txt_dir, file_name), 'r') as file:
            content = file.read()
            # Clean the content by replacing newline and tab characters with a space
            cleaned_content = content.replace('\n\t\t\t', ' ')
            contents.append(cleaned_content)

# Create a DataFrame with the titles and cleaned contents
df = pd.DataFrame({'Title': titles, 'Abstract': contents})

# Save the DataFrame to a CSV file
df.to_csv('output8000-9000.csv', index=False)
```
