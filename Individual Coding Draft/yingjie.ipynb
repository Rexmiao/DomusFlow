```python
import subprocess

def run_npm_commands():
    # Change directory to your project folder if necessary
    project_dir = '/path/to/your/project'
    try:
        # Installing npm dependencies
        print("Installing npm dependencies...")
        subprocess.run(["npm", "install"], check=True, cwd=project_dir)
        print("Dependencies installed successfully.")

        # Running the 'dev' script from package.json
        print("Starting development server...")
        subprocess.run(["npm", "run", "dev"], check=True, cwd=project_dir)
        print("Development server started.")
    except subprocess.CalledProcessError as e:
        print(f"Command failed with error: {e}")

# Execute the function
run_npm_commands()
```
